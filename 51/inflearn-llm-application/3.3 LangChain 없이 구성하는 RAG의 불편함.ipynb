```python
from docx import Document

document = Document('./tax.docx')

print(f'document == {dir(document)}')

for index, paragraph in enumerate(document.paragraphs):
    print(f'index == {index}, paragraph == {paragraph.text}')
    if index == 2:
        break

full_text = ''
for paragraph in document.paragraphs:
    full_text += paragraph.text + '\n'
```

```python
import tiktoken

encoder = tiktoken.encoding_for_model('gpt-4o')
encoding = encoder.encode(full_text)
decoded = encoder.decode(encoding)
```

```python
def split_text(text, chunk_size):
    total_encoding = encoder.encode(text)
    total_token_count = len(total_encoding)
    text_list = []
    for i in range(0, total_token_count, chunk_size):
        chunk = total_encoding[i:i+chunk_size]
        decoded_chunk = encoder.decode(chunk)
        text_list.append(decoded_chunk)

    return text_list
```

```python
chunk_list = split_text(full_text, 1500)
```

```python
import chromadb

chroma_client = chromadb.Client()
```

```python
import os
from dotenv import load_dotenv
from chromadb.utils.embedding_functions import OpenAIEmbeddingFunction

load_dotenv()

openai_api_key = os.getenv('OPENAI_API_KEY')
openai_embedding = OpenAIEmbeddingFunction(api_key=openai_api_key, model_name='text-embedding-3-large')
```

```python
collection_name = 'tax_collection'
tax_collection = chroma_client.get_or_create_collection(
    name=collection_name,
    embedding_function=openai_embedding
)
```

```python
id_list = []
for index in range(len(chunk_list)):
    id_list.append(f'{index}')
```

```python
len(id_list), len(chunk_list)
```

```python
tax_collection.add(documents=chunk_list, ids=id_list)
```

```python
query = '연봉 5천만원인 직장인의 소득세는 얼마인가요?'
retrieved_doc = tax_collection.query(query_texts=query)
```

```python
retrieved_doc
```

```python
from openai import OpenAI
client = OpenAI()

response = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "system", "content": f"당신은 한국의 소득세 전문가 입니다. 아래 내용을 참고해서 사용자의 질문에 답변해주세요 {retrieved_doc['documents'][0]}"},
    {"role": "user", "content": query}
  ]
)
```
